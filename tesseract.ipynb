```python
images = '../input/practise-data/screenshots Funnel 2'
for image in glob.glob(images + '/*.png'):
    img = cv2.imread(image)
    img = cv2.resize(img,(2000,2000))
    text = pytesseract.image_to_string(img)
#     print(text)
    b = word_tokenize(text)
    ct = 0

    for i in range(len(b)-2):
        if  "Insert Text Here" in b[i]+" "+ b[i+1]+" " + b[i+2]:
            ct = ct+1
    print('Title of the slide is: ',b[0],'Name of the image is :',os.path.split(image)[1],'Node count is :',ct)
```
